```python
import nltk
import numpy as np
import bs4 as bs
import urllib.request
import re
```

```python
raw_html=urllib.request.urlopen('https://en.wikipedia.org/wiki/Natural_language_processing')
raw_html
```

```python
raw_html=raw_html.read()
print(raw_html)
```

```python
atricle_html=bs.BeautifulSoup(raw_html,'lxml')
atricle_html
```

```python
article_paragraphs = atricle_html.find_all('p')#converting data to paragraphs
article_paragraphs 
```

```python
article_texts=''
for para in article_paragraphs:
    article_texts+=para.text
article_texts
#converting into single para
```

```python
w_token = nltk.word_tokenize(article_texts)
print(w_token)
len(w_token)
```

```python
words_no_punc = []
for w in w_token:
    if w.isalpha():
        words_no_punc.append(w)
print(words_no_punc)
len(words_no_punc)
```

```python
from nltk.corpus import stopwords
stopwords = stopwords.words('english')
print(stopwords)
```

```python
clean_words=[]
for w in words_no_punc:
    if w not in stopwords:
        clean_words.append(w)
clean_words 
```

```python
wordfreq={}
for token in clean_words:
        if token not in wordfreq:
            wordfreq[token]=1
        else:
            wordfreq[token]+=1
print(wordfreq)
len(wordfreq)    
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=433)
B_O_W = cv.fit_transform(clean_words).toarray()

print(cv.get_feature_names())
```

```python
print(cv.vocabulary_)
```
